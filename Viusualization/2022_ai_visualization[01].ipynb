```python
plt.figure( figsize = (15, 5) )  # 도화지 가져옴 (도화지 한 면이 만들어지는 것) > 변수 없이 쓰면 한면에 그래프 하나 다 씀
# fig = plt.figure( figsize = (20, 10) )  >  'fig='으로 변수 선언하면 도화지 면적 나눌 수 있음

x_data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y_data = [ x ** 2 for x in x_data ]


plt.plot(x_data, y_data,
         c = 'red',
         lw = '3',
         ls = '--',
         marker = 'o',
         ms = '7',
         mec = 'green',
         mew = '3',
         mfc = 'red')  

# data + option 지정 가능 
'''
- color : c
- line width : lw
- line style : ls
- marker shape : marker
- marker size : ms
- marker color : mfc
- marker edge color : mec
- marker edge width : mew
'''

plt.xlabel('x축')  # x - axis label
plt.ylabel('y축')  # y - axis label
plt.title('연습하는 lineplot')

plt.show()
plt.close()
```

### 시도별 전출입 인구수.xlsx 이용한 lineplot
---
[file load] 시도별 전출입 인구수.xlsx 

```python
pop_frm = pd.read_excel('../visualization_data/시도별 전출입 인구수.xlsx', header = 0)
pop_frm
```

```python
# front 에 있는 값으로 null 채워 넣기 : .fillna(method = 'ffill')
pop_frm = pop_frm.fillna(method = 'ffill')
pop_frm
```

```python
print('서울에서 다른 지역으로 이동한 데이터만 추출한다면?')
pop_frm.columns  # 전출지별, 전입지별

mask = (pop_frm['전출지별'] == '서울특별시') & (pop_frm['전입지별'] != '서울특별시' )
seoul_subset = pop_frm[mask]
seoul_subset
```

```python
print('전출지별 feature 를 삭제한다면?')
seoul_subset.drop('전출지별', axis=1, inplace=True)
```

```python
seoul_subset
```

```python
# 인덱스 재조정
seoul_subset.set_index('전입지별', inplace=True)
seoul_subset
```

```python
print('서울 -> 경기도로 이동한 인구 데이터만 선택한다면?')
datas = seoul_subset.loc['경기도']
print('type - ', type(datas))
print()
print('index - ', datas.index) # x axis
print()
print('values - ', datas.values)  # y axis
```

```python
print('경기도로 이동한 인구 데이터값을 이용한 lineplot - ')
print()

plt.figure( figsize = (15, 5) )

x_data = datas.index
y_data = datas.values


plt.plot(x_data, y_data,
         c = 'red',
         lw = '3',
         ls = '--',
         marker = 'o',
         ms = '7',
         mec = 'green',
         mew = '3',
         mfc = 'red')  


plt.xlabel('년도')                        # x - axis label
plt.ylabel('이동인구 수')                 # y - axis label
plt.title('서울 -> 경기 이동인구 수')

# 눈금 라벨 회전(x 축)
plt.xticks(rotation = 'vertical')

# 범례
plt.legend(labels = ['서울 -> 경기'], loc='best', fontsize = 15)  # loc='best' : 가장 best 한 위치를 너가 찾아라! (내가 지정해도 됨)

# y축의 범위 지정
plt.ylim(50000, 700000)

# 주석 표시
plt.annotate('인구 이동',
             xy = (20, 600000),       # 화살표의 머리 부분 (끝점)
             xytext = (2, 290000),    # text 보여지는 위치, 기준점, 시작점
             xycoords = 'data',
             fontsize = 15,
             rotation = -10,          # 주석 회전각도 설정 가능
             va = 'baseline',         # 상하정렬
             ha = 'center'            # text 좌우정렬 > 얘 해주게 되면 xytext 지정안해도 OK
            ) 

plt.show()
plt.close()
```

```python
print('subplot 이용한 시각화 - ')
# 하나의 면에 영역을 나누는 것

fig = plt.figure( figsize = (20, 10))    # 변수 지정해야 함

ax01 = fig.add_subplot(1, 2, 1)  # 1행 2열 짜리 중에 첫번째
ax02 = fig.add_subplot(1, 2, 2)  # 1행 2열 짜리 중에 두번째

ax01.plot(datas.index, datas.values)

ax02.plot(datas.index, datas.values)


plt.show()
plt.close()
```

충청남도, 경상북도, 강원도로 이동한 인구 데이터값만 선택해서 하나의 figure에 시각화 (1 figure , 3 plot) == plt.plot 3번 시행
