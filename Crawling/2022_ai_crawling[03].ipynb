#### XPATH
- XML의 문서의 노드나 노드셋을 선택하는 방법
- 경로연산자
- 노드이름 : 모든 노드를 선택
- / : 루트 노드부터 순서대로 탐색하여 선택
- // : 지정된 노드에서부터 순서대로 탐색
- . : 현재노드
- .. : 부모노드 
- @  : 속성노드 선택
- 예) 
- 노드 예시 <div name = 'test' , id='content'>
- driver.find_element_by_xpath('//div[@id='content']')

```python
path = './driver/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get('https://deal.11st.co.kr/browsing/DealAction.tmall?method=getShockingDealMain')


```

```python
print('css selector - ')
print()

li_tags = driver.find_elements_by_css_selector('#emergencyPrd > div > ul > li')

for li in li_tags :
    # print(li)
    # print('*' * 50)
    
    info = li.find_element_by_css_selector('div > a > div.prd_info > p > span.fs_16')
    print(info.text , '--' ,  info.get_attribute('innerHTML'))
    price = li.find_element_by_css_selector('div > a > div.prd_info > div > span.price_detail > strong')
    print(price.text , '--' ,  price.get_attribute('innerHTML'))
    print('*' * 50)
```

```python
print('xpath - ')
print()

li_tags = driver.find_elements_by_xpath('//*[@id="emergencyPrd"]/div/ul/li[position() <= 3 ]')


for li in li_tags :
    
    print( li.find_element_by_xpath('div/a/div[3]/p/span[2]').text )
    print( li.find_element_by_xpath('div/a/div[3]/div/span[2]/strong').text )
    
    # print( li )
    print('*' * 50)
    
    
```
